# Suppression de l'arrière-plan de l'image à l'aide de 'rembg'
Ce code accédera à un dossier contenant des images, supprimera l'arrière-plan de chaque image et enregistrera les nouvelles images dans un dossier différent.

```python
# Monter le lecteur là où se trouvent les dossiers d'imagesfrom google.colab import drive
drive.mount('/content/drive')
```

```python
# Importez les bibliothèques nécessairesfrom rembg import remove
from PIL import Image
import io
import glob
```

```python
# Cette boucle sur chaque image supprime l'arrière-plan et enregistre dans un nouveau répertoirefor file in files:
  input_path = file
  output_path = input_path.replace("Dataset", "Dataset_new")
  output_path = output_path.replace("JPG", "PNG")
  
  with open(input_path, 'rb') as i:
    with open(output_path, 'wb') as o:
        input = i.read()
        output = remove(input)
        o.write(output)
```
